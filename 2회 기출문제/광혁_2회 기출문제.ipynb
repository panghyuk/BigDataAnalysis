# 1-3 다음은 insurance 데이터 셋이다. charges 항목의 이상값의 합을 구하시오. (이상값은 평균에서 1.5 표준편차 이상인 값)

```python
import pandas as pd
```

```python
df = pd.read_csv("insurance.csv")
display(df)
df.info()
```

```python
charge_mean = df['charges'].mean()
charge_std = df['charges'].std()
top = charge_mean + 1.5 * charge_std
bottom = charge_mean - 1.5 * charge_std
```

```python
total_sum = 0
for val in df['charges']:
    if val > top:
        total_sum += val
    elif val < bottom:
        total_sum += val
```

```python
print(total_sum)
```

# 2 아래 E-commerce Shipping Data의 train set을 참고하여 고객이 주문한 물품의 정시 도착 여부를 예측하시오. (ID와 예측치를 csv 파일로 저장하여 제출)

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
```

```python
X_train = pd.read_csv("X_train.csv")
X_test = pd.read_csv("X_test.csv")
y_train = pd.read_csv("y_train.csv")
```

```python
X_train.info()
```

```python
X_cat = X_train[['Warehouse_block','Mode_of_Shipment','Product_importance','Gender']]
X_num = X_train[['Customer_care_calls','Customer_rating','Cost_of_the_Product','Prior_purchases','Discount_offered','Weight_in_gms']]
X_cat = pd.get_dummies(X_cat)
```

```python
test_cat = X_test[['Warehouse_block','Mode_of_Shipment','Product_importance','Gender']]
test_num = X_test[['Customer_care_calls','Customer_rating','Cost_of_the_Product','Prior_purchases','Discount_offered','Weight_in_gms']]
test_cat = pd.get_dummies(test_cat)
```

```python
# 
X_cat, test_cat = X_cat.align(test_cat, join = 'inner', axis = 1)
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_num)
test_scaled = scaler.transform(test_num)
```

```python
# 
X_final = pd.concat([pd.DataFrame(X_scaled), X_cat], axis = 1)
test_final = pd.concat([pd.DataFrame(test_scaled), test_cat], axis = 1)
```

```python
y_fin = y_train['Reached.on.Time_Y.N']
y_fin
```

```python
model = LogisticRegression()
model.fit(X_final,y_fin)
```

```python
pred_test = model.predict_proba(test_final)
pred_test
```

```python
pred_test_prob = pd.DataFrame(pred_test[:,0],columns = ['predict_prob'])
pred_test_prob
```

```python
final_pred = pd.concat([X_test['ID'],pred_test_prob],axis = 1)
final_pred.to_csv("20220616.csv",index = False)
```
