# 02 Filtering & Sorting

```python
# 20
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(url)
ans = type(df)
ans
```

```python
# 21 quantity 칼럼 값이 3인 데이터를 추출하여 첫 5행 출력
ans = df.loc[df['quantity'] == 3].head()
ans
```

```python
ans = df[df['quantity'] == 3].head()
ans
```

```python
# 22 drop true 옵션을 통해 기존 인덱스를 버리고 재배열
ans = df.loc[df['quantity'] == 3].head().reset_index(drop = True)
ans
```

```python
ans = df[df['quantity'] == 3].head().reset_index(drop = True)
ans
```

```python
# 23
ans = df[['quantity','item_price']]
ans
```

```python
# 24 item_price 칼럼의 달러 표시 문자 제거 후 float 타입으로 변환하여 new_price에 저장
df['new_price'] = df['item_price'].str[1:].astype('float')
ans = df['new_price'].head()
ans
```

```python
# 25
ans = len(df.loc[df['new_price'] <= 5])
ans
```

```python
# 26
ans = df.loc[df.item_name == 'Chicken Salad Bowl'].reset_index(drop = True).head(3)
ans
```

```python
ans = df[df['item_name'] == 'Chicken Salad Bowl'].reset_index(drop = True).head(3)
ans
```

```python
# 27 두 조건을 만족하는 데이터 프레임
ans = df[(df['new_price'] <= 9) & (df['item_name'] == 'Chicken Salad Bowl')].head()
ans
```

```python
ans = df.loc[(df.new_price <= 9) & (df.item_name == 'Chicken Salad Bowl')].head()
ans
```

```python
# 28 칼럼 값에 따른 오름차순 정렬
ans = df.sort_values('new_price').reset_index(drop = True).head()
ans
```

```python
# 29 칼럼 값에서 특정한 문자열을 포함하는 경우
ans = df.loc[df['item_name'].str.contains('Chips')].head()
ans
```

```python
# 30
ans = df.loc[:,::2]
ans
```

```python
# 31
ans = df.sort_values(by = 'new_price',ascending = False).reset_index(drop = True).head()
ans
```

```python
# 32
ans = df[(df['item_name'] == 'Steak Salad')|(df['item_name'] == 'Bowl')]
ans
```

```python
# 33 중복행 제거(첫번째 케이스)
ans = df[(df['item_name'] == 'Steak Salad')|(df['item_name'] == 'Bowl')]
ans = ans.drop_duplicates('item_name')
ans
```

```python
# 34 중복행 제거(마지막 케이스)
ans = df[(df['item_name'] == 'Steak Salad')|(df['item_name'] == 'Bowl')]
ans = ans.drop_duplicates('item_name',keep = "last")
ans
```

```python
# 35
ans = df[df['new_price'] > df['new_price'].mean()]
ans
```

```python
# 36 특정 칼럼의 값 변경
df.loc[df['item_name'] == 'Izze','item_name'] = 'Fizzy Lizzy'
ans = df.head()
ans
```

```python
# 37
ans = df['choice_description'].isnull().sum()
ans
```

```python
# 38 NaN 값 대체
df.loc[df['choice_description'].isnull(),'choice_description'] = 'No Data'
ans = df.head()
ans
```

```python
# 39 특정 문자열 포함한 경우
ans = df.loc[df['choice_description'].str.contains('Black')].head()
ans
```

```python
# 40 특정 문자열 포함하지 않는 경우
ans = len(df.loc[~ df['choice_description'].str.contains('Vegetables')])
ans
```

```python
# 41 특정 문자열로 시작
ans = df.loc[df['item_name'].str.startswith('N')].head()
ans
```

```python
# 42 문자열 길이
ans = df.loc[df['item_name'].str.len() >= 15]
ans
```

```python
# 43 리스트 내에 있는 값과 비교(포함되는지?)
lst =[1.69, 2.39, 3.39, 4.45, 9.25, 10.98, 11.75, 16.98]
ans = df.loc[df['new_price'].isin(lst)]
display(ans.head())
print(len(ans))
```

# 03 Grouping

```python
# 44
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'
df = pd.read_csv(url)
df.head()
```

```python
# 45 빈도수 구하기 & 정렬/Group By -> 자동 정렬
ans = df.groupby("host_name").size().head()
ans
```

```python
ans = df['host_name'].value_counts().sort_index().head()
ans
```
