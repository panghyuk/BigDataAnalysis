# 04 Apply, Map

```python
# 56
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
df = pd.read_csv(url,index_col = 0)
display(df)
df.shape
```

```python
# 57 map 함수
dic = {
    'Unknown':'N',
    'Less than $40K':'a',
    '$40K - $60K':'b',
    '$60K - $80K':'c',
    '$80K - $120K':'d',
    '$120K +':'e',
}

df['newIncome'] = df['Income_Category'].map(lambda x: dic[x])
ans = df['newIncome'].head()
ans
```

```python
# 58 apply 함수
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'
    
df['newIncome'] = df['Income_Category'].apply(changeCategory)
ans = df['newIncome'].head()
ans
```

```python
# 59
df['AgeState'] = df['Customer_Age'].map(lambda x: (x // 10) * 10)
ans = df['AgeState'].value_counts().sort_index()
ans
```

```python
# 60
df['newEduLevel'] = df['Education_Level'].map(lambda x: 1 if 'Graduate' in x else 0)
ans = df['newEduLevel'].value_counts()
ans
```

```python
# 61
df['newLimit'] = df['Credit_Limit'].map(lambda x: 1 if x >= 4500 else 0)
ans = df['newLimit'].value_counts()
ans
```

```python
# 62
def check(x):
    if x['Marital_Status'] == "Married" and x['Card_Category'] == 'Platinum':
        return 1
    else:
        return 0

df['newState'] = df.apply(check, axis = 1)
ans = df['newState'].value_counts()
ans
```

```python
# 63
def changesex(x):
    if x == "M":
        return 'male'
    else:
        return 'female'
    
df['Gender'] = df['Gender'].apply(changesex)
ans = df['Gender'].value_counts()
ans
```

```python
# 67
def fix_century(x):
    import datetime
    if x.year > 2060:
        year = x.year - 100
    else:
        year = x.year
    
    return pd.to_datetime(datetime.date(year, x.month, x.day))


df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)

Ans = df
Ans
```

```python
# 76
url = "https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv"
df = pd.read_csv(url)
df
```

```python
# 76
def change_date(x):
    import datetime
    h = x.split(":")[1]
    d = x.split(":")[0]
    
    if h == '24':
        h = "00:00:00"
        
        final_date = pd.to_datetime(d+" "+h) + datetime.timedelta(days = 1)
    else:
        h = h + ":00:00"
        final_date = pd.to_datetime(d+" "+h)
    
    return final_date
```

```python
df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df.head()
```
